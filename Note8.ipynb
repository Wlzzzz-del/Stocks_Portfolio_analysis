#  Visualize Time Series Data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import matplotlib.dates as dates
```

```python
mcdon = pd.read_csv('mcdonalds.csv',index_col='Date',parse_dates=True)
```

```python
mcdon.plot()
```

```python
mcdon['Adj. Close'].plot()
```

```python
mcdon['Adj. Volume'].plot(figsize=(12,8))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=[20,60])
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2007-05-01'],ylim=[0,40])
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx,stock,'g-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)

fig.autofmt_xdate()
```

```python
#Major Axis,Minor Axis
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx,stock,'-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)

#Major
#1. Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
#2. Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

#Minor
#1. Locating
ax.xaxis.set_minor_locator(dates.WeekdayLocator())
#2. Formatting
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))


fig.autofmt_xdate()
```
