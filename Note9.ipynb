```python
df = pd.read_csv('walmart_stock.csv',index_col = 'Date',parse_dates = True)
df.head()
```

```python
df['Open'].plot(figsize = (16,6))
```

```python
#7-day rolling mean
df.rolling(7).mean()
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window = 30).mean()
df.tail(20)
```

```python
df[['Close','Close 30 Day MA']].plot(figsize = (16,6))
```

```python
#Expanding:从数据开始一直到当下时间点
```

```python
df['Close'].expanding(min_periods = 1).mean().plot(figsize = (16,6))
```

```python
df['Close 20 Day MA'] = df['Close'].rolling(window = 20).mean()
df['Upper'] = df['Close 20 Day MA'] + 2*df['Close'].rolling(window = 20).std()
df['Lower'] = df['Close 20 Day MA'] - 2*df['Close'].rolling(window = 20).std()
df[['Close','Close 20 Day MA','Upper','Lower']].plot(figsize = (16,6))
```
